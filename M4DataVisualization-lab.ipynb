## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


```python
query = "SELECT * FROM master"
df = pd.read_sql_query(query, conn)
```

Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt
plt.hist(df['ConvertedComp'].dropna(), bins=25, edgecolor='black')
plt.title('Converted Compensation Histogram')
plt.xlabel('Converted Compensation (USD)')
plt.ylabel('Frequency')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
plt.boxplot(df['Age'].dropna(), vert=False)
plt.title('Boxplot - Age')
plt.xlabel('Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
plt.scatter(df['Age'], df['WorkWeekHrs'])
plt.title('Age vs. Work Hours per Week')
plt.xlabel('Age')
plt.ylabel('Work Hours/Week')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=df['Age'])
plt.title('Work Hours per Week vs. Code Review Hours')
plt.xlabel('Work Hours/Week')
plt.ylabel('Code Review Hours')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
query_ddny = "SELECT * FROM DatabaseDesireNextYear"
df_ddny = pd.read_sql_query(query_ddny, conn)
```

```python
count_db = df_ddny['DatabaseDesireNextYear'].value_counts()
top_5_db = count_db.nlargest(5)
plt.pie(top_5_db, labels=top_5_db.index, autopct='%1.1f%%')
plt.title("Top 5 Databases Respondents Wish to Learn Next Year")
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
age_df = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

median_WorkWeekHrs = age_df['WorkWeekHrs'].median()
median_CodeRevHrs = age_df['CodeRevHrs'].median()

categories = ['Age 30-35']
work_week_data = [median_WorkWeekHrs]
code_rev_data = [median_CodeRevHrs]

plt.bar(categories, work_week_data, label='WorkWeekHrs', color='skyblue')
plt.bar(categories, code_rev_data, bottom=work_week_data, label='CodeRevHrs', color='salmon')

plt.xlabel('Age Group')
plt.ylabel('Median Hours')
plt.title('Median Work Week Hours and Code Review Hours - Age 30-35')
plt.legend()

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
age45_60_df = df[(df['Age'] >= 45) & (df['Age'] <= 60)]
median_comp_age = age45_60_df.groupby('Age')['ConvertedComp'].median()

plt.plot(median_comp_age.index, median_comp_age.values, marker='o')
plt.title('Median Median Converted Compensation - Age 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation (USD)')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
main_branch_count = df['MainBranch'].value_counts()
main_branch_count.plot(kind='barh', color='skyblue')

plt.xlabel('Number of Respondents')
plt.ylabel('Main Branch')
plt.title('Respondents by Main Branch')
plt.show()
```
